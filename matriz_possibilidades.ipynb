```python
import pandas as pd
```

```python
matriz_de_possibilidades = []
total_linhas = 2**21
```

```python
for l in range(21):
  a = 2**l
  vet = []
  repeticoes = int( total_linhas / (a*2) )

  for k in range(repeticoes):
    for i in range(a):
      vet.append(0);
    for i in range(a):
      vet.append(1);

  matriz_de_possibilidades.append(vet)

df_possibilidades = pd.DataFrame(matriz_de_possibilidades).transpose()
df_possibilidades = df_possibilidades.drop(0).reset_index(drop=True)
df_possibilidades = df_possibilidades.sample(frac=1).reset_index(drop=True)
df_possibilidades
```

```python
df_possibilidades.to_csv('dados/matriz_possibilidades.csv', index=False)
```
